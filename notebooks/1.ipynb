```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
import numpy as np
```

```python

# Filtering the dataset to include only rows where 'number_of_bedrooms' is not null
data_for_model = Glenroy[['total_rooms', 'property_type', 'sale_price', 'number_of_bedrooms']].dropna()

# Separating the features and target variable
X = data_for_model[['total_rooms', 'property_type', 'sale_price']]
y = data_for_model['number_of_bedrooms']
```

```python

# Encoding the categorical data
categorical_features = ['property_type']
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
```

```python

# Applying transformations to the dataset
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_features)
    ], remainder='passthrough')

# Define the model
model = make_pipeline(preprocessor, RandomForestRegressor(n_estimators=100, random_state=0))
```

```python

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train the model
model.fit(X_train, y_train)

# Predicting the 'number_of_bedrooms' for the rows where it's NaN
predict_data = Glenroy[Glenroy['number_of_bedrooms'].isnull()][['total_rooms', 'property_type', 'sale_price']]
predicted_bedrooms = model.predict(predict_data)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Evaluate the model
y_pred = model.predict(X_test)

# Convert predictions to integers (if using RandomForestRegressor)
y_pred = np.round(y_pred).astype(int)

# Evaluation metrics
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Generate classification report
report = classification_report(y_test, y_pred, output_dict=True)
report
```

```python
# Plotting the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='g')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python

# Generate classification report
report = classification_report(y_test, y_pred, output_dict=True)

# Convert the report to a DataFrame
report_df = pd.DataFrame(report).transpose()

# Plotting the classification report
plt.figure(figsize=(10, 5))
sns.heatmap(report_df.iloc[:-1, :].T, annot=True, cmap="Blues")
plt.title("Classification Report Heatmap")
plt.show()
```

```python

# Fill the NaN values in 'number_of_bedrooms' with the predictions
Glenroy.loc[Glenroy['number_of_bedrooms'].isnull(), 'number_of_bedrooms'] = predicted_bedrooms

# Rounding the predicted values for 'Bathroom' to the nearest integer
Glenroy['number_of_bedrooms'] = Glenroy['number_of_bedrooms'].apply(np.round)

# Ensure that the 'Bathroom' column is of integer type
Glenroy['number_of_bedrooms'] = Glenroy['number_of_bedrooms'].astype(int)

# Check if there are any null values remaining in 'number_of_bedrooms'
remaining_nulls = Glenroy['number_of_bedrooms'].isnull().sum()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# One-hot encode the 'property_type' column
encoder = OneHotEncoder(sparse=False)
property_type_encoded = encoder.fit_transform(Glenroy_tree[['property_type']])

# Create a new DataFrame with encoded property_type
encoded_columns = pd.DataFrame(property_type_encoded, columns=encoder.get_feature_names_out(['property_type']))
data_encoded = pd.concat([Glenroy_tree.drop('property_type', axis=1), encoded_columns], axis=1)

# Split the data into features (X) and target (y)
X = data_encoded[['total_rooms', 'sale_price'] + list(encoded_columns.columns)]
y = data_encoded['number_of_bedrooms']

# Split the data into training and testing sets
# Training set: Rows with non-null 'number_of_bedrooms'
# Testing set: Rows with null 'number_of_bedrooms'
X_train = X[y.notnull()]
y_train = y[y.notnull()]
X_test = X[y.isnull()]

# Train a Decision Tree model
tree = DecisionTreeRegressor()
tree.fit(X_train, y_train)

# Predict the missing values of 'number_of_bedrooms' in the test set
predicted_bedrooms = tree.predict(X_test)

# Add the predictions to the original dataset
Glenroy_tree.loc[y.isnull(), 'number_of_bedrooms'] = predicted_bedrooms

# Convert the predicted 'number_of_bedrooms' to integers
Glenroy_tree['number_of_bedrooms'] = Glenroy_tree['number_of_bedrooms'].round().astype(int)

# Display the first few rows of the dataframe with the integer 'number_of_bedrooms'
Glenroy_tree.head()

```
